```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def numerical_diff(f, x):
    h = 1e-4  # 0.0001
    return (f(x+h) - f(x-h)) / (2*h)

def tangent_line(f, x):
    d = numerical_diff(f, x)
    y = f(x) - d*x
    return lambda t: d*t + y

def function_1(x):
    return x ** 2

def function_2(x):
    return x**3 + x**2 + x

def function_3(x):
    return x * np.sin(x*x*x + 3)

def function_4(x):
    return np.cos(np.sqrt(np.e**x + 1) / 2)

x = np.arange(0.0, 20.0, 0.1)  # 0에서 20까지 0.1 간격의 배열 x를 만든다.

y1 = function_1(x)
tf = tangent_line(function_1, 10)
y2 = tf(x)

y3 = function_1(x)
tf = tangent_line(function_2, 10)
y4 = tf(x)

y5 = function_1(x)
tf = tangent_line(function_3, 10)
y6 = tf(x)

y7 = function_1(x)
tf = tangent_line(function_4, 10)
y8 = tf(x)

plt.xlabel('x') 
plt.ylabel('f(x)')
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.plot(x, y5)
plt.plot(x, y6)
plt.plot(x, y7)
plt.plot(x, y8)
plt.show()
```
